```python
import os
```

```python
class Result():
    def __init__(self, test1,test2,test3,name,fn1,fn2,fn3):
        self.test1 = test1
        self.test2 = test2
        self.test3 = test3
        self.name = name
        self.fn1 = fn1
        self.fn2 = fn2
        self.fn3 = fn3
        self.summary=None
        self.weight=None
        self.mae=None

    def set_summary(self,summary):
        self.summary=summary

    def set_weight(self,weight):
        self.weight=weight

#read all files in the folder
root="./predictions/37orwro0/"
names=["Test1_preds.txt","Test2_preds.txt","Test3_preds.txt"]
results=[]
files = os.listdir(root)
for f in files:
    test1=[
    ]
    test2=[]
    test3=[]
    fn1=[]
    fn2=[]
    fn3=[]
    #check that the folder contains the 3 files
    if not os.path.exists(os.path.join(root,f,names[0])):
        continue
    if not os.path.exists(os.path.join(root,f,names[1])):
        continue
    if not os.path.exists(os.path.join(root,f,names[2])):
        continue
    
    with open(os.path.join(root,f,names[0])) as f1:
        for line in f1:
            #C1/3-1-2-submit-00000.mp4,2.9382266998291016
            l=line.split(",")
            fn1.append(l[0])
            test1.append(float(l[1]))

    print(os.path.join(root,f,names[1]))

    with open(os.path.join(root,f,names[1])) as f2:
        for line in f2:
            #C1/3-1-2-submit-00000.mp4,2.9382266998291016

            l=line.split(",")
            fn2.append(l[0])
            test2.append(float(l[1]))

    with open(os.path.join(root,f,names[2])) as f3:
        for line in f3:
            #C1/3-1-2-submit-00000.mp4,2.9382266998291016

            l=line.split(",")
            fn3.append(l[0])
            test3.append(float(l[1]))


    result=Result(test1,test2,test3,f,fn1,fn2,fn3)
    results.append(result)

root="./eva_models/37orwro0"
names=["Test1_preds.txt","Test2_preds.txt","Test3_preds.txt"]
submition=[]
files = os.listdir(root)
for f in files:
    test1=[
    ]
    test2=[]
    test3=[]
    fn1=[]
    fn2=[]
    fn3=[]
    #check that the folder contains the 3 files
    if not os.path.exists(os.path.join(root,f,names[0])):
        continue
    if not os.path.exists(os.path.join(root,f,names[1])):
        continue
    if not os.path.exists(os.path.join(root,f,names[2])):
        continue
    
    with open(os.path.join(root,f,names[0])) as f1:
        for line in f1:
            #C1/3-1-2-submit-00000.mp4,2.9382266998291016
            l=line.split(",")
            fn1.append(l[0])
            test1.append(float(l[1]))

    print(os.path.join(root,f,names[1]))

    with open(os.path.join(root,f,names[1])) as f2:
        for line in f2:
            #C1/3-1-2-submit-00000.mp4,2.9382266998291016

            l=line.split(",")
            fn2.append(l[0])
            test2.append(float(l[1]))

    with open(os.path.join(root,f,names[2])) as f3:
        for line in f3:
            #C1/3-1-2-submit-00000.mp4,2.9382266998291016

            l=line.split(",")
            fn3.append(l[0])
            test3.append(float(l[1]))


    result=Result(test1,test2,test3,f,fn1,fn2,fn3)
    submition.append(result)

print(len(results))
print(len(submition))
```

```python
#calculate avg of all res
test1=[]
test2=[]
test3=[]
for r in results:
    test1.append(r.test1)
    test2.append(r.test2)
    test3.append(r.test3)

test1=np.array(test1)
test2=np.array(test2)
test3=np.array(test3)

avg1=np.mean(test1,axis=0)
avg2=np.mean(test2,axis=0)
avg3=np.mean(test3,axis=0)

plt.figure()
plt.plot(avg1[:100],label="avg1")
plt.plot(submition[0].test1[:100],label="submition",c="lime")
plt.legend()
plt.show()

plt.figure()
plt.plot(avg2[:100],label="avg2")
plt.plot(submition[0].test2[:100],label="submition",c="lime")
plt.legend()
plt.show()

plt.plot(avg3[:100],label="avg2")
plt.plot(submition[0].test3[:100],label="submition",c="lime")
plt.legend()
plt.show()



    
```
